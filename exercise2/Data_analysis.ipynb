# Scalability Study
## Fixed cores and variable matrix size

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ['m', 'k', 'n', 'time', 'gflops']
MKL_close= pd.read_csv("./fxed_cores" + '/close_cores/mkl' + "_f" + '.csv', names=column_names, header=0)

MKL_close = MKL_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')
```

```python
def plot_results(fixed_quantity, node_kind, precision, tpp):
    res_path = "./fixed_" + fixed_quantity + "/" + node_kind + "/"
    precision = "_" + precision
    
    linetype = ""
    column_names = ['m', 'k', 'n', 'time', 'gflops']
    
    
    
    MKL_close= pd.read_csv(res_path + '/close_cores/mkl' + precision + '.csv', names=column_names, header=0)
    
    MKL_close = MKL_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')

    OPENBLAS_close= pd.read_csv(res_path + '/close_cores/openblas' + precision + '.csv', names=column_names, header=0)
    
    OPENBLAS_close = OPENBLAS_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')

    BLIS_close = pd.read_csv(res_path + '/close_cores/blis' + precision + '.csv', names=column_names, header=0)
    
    BLIS_close = BLIS_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')
    
    

    MKL_spread = pd.read_csv(res_path + '/spread_cores/mkl' + precision + '.csv', names=column_names, header=0)
    
    MKL_spread = MKL_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')

    OPENBLAS_spread = pd.read_csv(res_path + '/spread_cores/openblas' + precision + '.csv', names=column_names, header=0)
    
    OPENBLAS_spread = OPENBLAS_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')

    BLIS_spread = pd.read_csv(res_path + 'spread_cores/blis' + precision + '.csv', names=column_names, header=0)
    
    BLIS_spread = BLIS_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')


    plt.plot(MKL_close['m'], MKL_close['gflops'], linetype, label = "MKL Close", color = 'deepskyblue')
    plt.plot(OPENBLAS_close['m'], OPENBLAS_close['gflops'], linetype, label = "OPENBLAS Close", color = 'dodgerblue')
    plt.plot(BLIS_close['m'], BLIS_close['gflops'], linetype, label = "BLIS Close", color = 'blue')

    plt.plot(MKL_spread['m'], MKL_spread['gflops'], linetype, label = "MKL Spread", color = 'springgreen')
    plt.plot(OPENBLAS_spread['m'], OPENBLAS_spread['gflops'], linetype, label = "OPENBLAS Spread", color = 'limegreen')
    plt.plot(BLIS_spread['m'], BLIS_spread['gflops'], linetype, label = "BLIS Spread", color = 'forestgreen')

    plt.plot(BLIS_close['m'], [tpp] * BLIS_close['m'].size, label = "TPP", color = 'pink')
    plt.xlabel("Matrix dimension")
    plt.ylabel("Operations per seconds [Gflops/s]")
    plt.legend()
    plt.savefig("fixed_" + fixed_quantity + "_" + node_kind + precision + ".png")
    plt.close()
```

```python
fixed_quantity = "matrix"

node_kind = "THIN"

precision = "f"

res_path = "./fixed_" + fixed_quantity + "/" + node_kind + "/"
precision = "_" + precision

linetype = ""
column_names = ['m', 'k', 'n', 'time', 'gflops']



MKL_close= pd.read_csv(res_path + '/close_cores/mkl' + precision + '.csv', names=column_names, header=0)

MKL_close = MKL_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')

OPENBLAS_close= pd.read_csv(res_path + '/close_cores/openblas' + precision + '.csv', names=column_names, header=0)

OPENBLAS_close = OPENBLAS_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')

BLIS_close = pd.read_csv(res_path + '/close_cores/blis' + precision + '.csv', names=column_names, header=0)

BLIS_close = BLIS_close.groupby("m")[['m', 'gflops', 'time']].transform('mean')



MKL_spread = pd.read_csv(res_path + '/spread_cores/mkl' + precision + '.csv', names=column_names, header=0)

MKL_spread = MKL_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')

OPENBLAS_spread = pd.read_csv(res_path + '/spread_cores/openblas' + precision + '.csv', names=column_names, header=0)

OPENBLAS_spread = OPENBLAS_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')

BLIS_spread = pd.read_csv(res_path + 'spread_cores/blis' + precision + '.csv', names=column_names, header=0)

BLIS_spread = BLIS_spread.groupby("m")[['m', 'gflops', 'time']].transform('mean')


plt.plot(MKL_close['m'], MKL_close['gflops'], linetype, label = "MKL Close", color = 'deepskyblue')
plt.plot(OPENBLAS_close['m'], OPENBLAS_close['gflops'], linetype, label = "OPENBLAS Close", color = 'dodgerblue')
plt.plot(BLIS_close['m'], BLIS_close['gflops'], linetype, label = "BLIS Close", color = 'blue')

plt.plot(MKL_spread['m'], MKL_spread['gflops'], linetype, label = "MKL Spread", color = 'springgreen')
plt.plot(OPENBLAS_spread['m'], OPENBLAS_spread['gflops'], linetype, label = "OPENBLAS Spread", color = 'limegreen')
plt.plot(BLIS_spread['m'], BLIS_spread['gflops'], linetype, label = "BLIS Spread", color = 'forestgreen')

plt.plot(BLIS_close['m'], [tpp] * BLIS_close['m'].size, label = "TPP", color = 'pink')
plt.xlabel("Matrix dimension")
plt.ylabel("Operations per seconds [Gflops/s]")
plt.legend()
plt.savefig("fixed_" + fixed_quantity + "_" + node_kind + precision + ".png")
plt.close()
```

```python
plot_results(fixed_quantity="cores", node_kind="THIN", precision="f", tpp=1997 * 0.5 * 2)
plot_results(fixed_quantity="cores", node_kind="THIN", precision="d", tpp=1997 * 0.5)

plot_results(fixed_quantity="cores", node_kind="EPYC", precision="f", tpp=2.60 * 16 * 64 * 2)
plot_results(fixed_quantity="cores", node_kind="EPYC", precision="d", tpp=2.60 * 16 * 64)
```
